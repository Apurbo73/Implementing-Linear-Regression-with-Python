```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/nasdaq100.csv')
df.head(5)
```

```python
df = pd.read_csv('/content/nasdaq100.csv',sep=';')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Date'])
df
```

```python
plt.scatter(df['Starting (USD)'], df['Ending (USD)'])
plt.xlabel('Starting (USD)')
plt.ylabel ('Ending (USD)')
plt.title ('Nasdaq 100 Stock Prices')
```

```python
x=df.drop(columns=['Ending (USD)'])

x.head()
```

```python
y= df['Ending (USD)']
y.head()
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Starting (USD)'], df['Ending (USD)'])
plt.xlabel('Starting (USD)')
plt.ylabel ('Ending (USD)')
plt.title ('Nasdaq 100 Stock Prices')
```

```python
reg.fit(x,y) # Training the model
m= reg.coef_
m
```

```python
c=reg.intercept_
c
```

```python
reg.predict([[16700]])
```

```python
df['Predicted Y'] = reg.predict(x)
df.head()
```

```python
plt.plot(x,reg.predict(x),color='green')
plt.scatter(x.mean(),y.mean(),color='red')
plt.scatter(df['Starting (USD)'], df['Ending (USD)'])
plt.xlabel('Starting (USD)')
plt.ylabel ('Ending (USD)')
plt.title ('Nasdaq 100 Stock Prices')
```
